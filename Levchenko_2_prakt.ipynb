```python
import pandas as pd
data = pd.read_csv("COVID-19-in-Ukraine-from-April.csv")
data.head()
```

# Графік 1

```python
population_by_continent = {
    'Deaths': data['deaths'],
    'Recovered': data['recovered'],
    'Existing': data['existing'],
}
fig, ax = plt.subplots(figsize = (15,10))
ax.stackplot(data['confirmed'], population_by_continent.values(),
             labels=population_by_continent.keys(), alpha=0.8)
ax.legend(loc='upper left')
ax.set_title('COVID-19 in Ukraine from April')
ax.set_xlabel('Confirmed')
ax.set_ylabel('Data')
plt.ylim(0,250000)
plt.show()
```

# Графік 2

```python
from matplotlib import patheffects

fig, ax = plt.subplots(figsize=(6, 6))
x = data['deaths']
y = data['recovered']
ax.set_title('Comprasion deaths and recovered  from COVID-19 in Ukraine from April')
ax.plot(x, label="Deaths",
        path_effects=[patheffects.withTickedStroke(spacing=7, angle=135)])
ax.plot(y, label="Recovered", path_effects=[patheffects.withTickedStroke()])
ax.legend()

plt.show()

```

# Графік 3

```python
x = data['n_existing']
y = data['n_confirmed']
z = data['n_deaths']
fig = plt.figure(figsize=(10,5))
ax = fig.add_subplot(111, projection='3d')
ax.set_title('Comprasion existing and deaths from COVID-19 in Ukraine from April')
ax.scatter(y,x, color ='b')
ax.scatter(y,z, color ='y')
```
