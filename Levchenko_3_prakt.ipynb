# Графік 1

```python
sns.displot(
    data=titanic,
    x="age", hue="pclass",
    kind="kde", multiple="fill", clip=(0, None), palette = "ch:start=.5,rot=-.75"
)
plt.title('Comprasion')
```

# Графік 3

```python
import numpy as np

n = 10000
mean = [0, 0]
cov = [(3, .5), (.5, .3)]
rng = np.random.RandomState(0)
x, y = rng.multivariate_normal(mean, cov, n).T

f, ax = plt.subplots(figsize=(6, 6))
sns.scatterplot(x=x, y=y, s=5, color=".15")
sns.histplot(x=x, y=y, bins=50, pthresh=.1, cmap="Spectral")
sns.kdeplot(x=x, y=y, levels=5, color="y", linewidths=1)
```
